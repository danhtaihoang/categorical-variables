# Network inference from time series of categorical variables

```python
import sys,os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import inference
```

```python
# setting parameter:
np.random.seed(1)

n = 20 # number of positions
m = 3 # number of values at each position
l = int(4*((n*m)**2)) # number of samples
g = 1.
```

```python
w0 = inference.generate_interactions(n,m,g)
```

```python
s = inference.generate_sequences(w0,n,m,l)
```

## Large sample sizes

```python
w1 = inference.fit_additive(s,n,m)
```

```python
plt.scatter(w0,w1)
plt.plot([-0.5,0.5],[-0.5,0.5],'r--')
```

```python
mse = ((w0-w1)**2).mean()
print(mse)
```

plt.scatter(w0,w1)
plt.plot([-0.5,0.5],[-0.5,0.5],'r--')

```python
w2 = inference.fit_multiplicative(s,n,m)
```

```python
plt.scatter(w0,w2)
plt.plot([-0.5,0.5],[-0.5,0.5],'r--')
```

```python
mse = ((w0-w2)**2).mean()
print(mse)
```

## Small sample sizes

```python
l2 = int(0.5*(n*m)**2)
```

### Additive update

```python
w1 = inference.fit_additive(s[:l2],n,m)
```

```python
plt.scatter(w0,w1)
plt.plot([-0.5,0.5],[-0.5,0.5],'r--')
mse = ((w0-w1)**2).mean()
print('MSE by additive update:',mse)
```

### Multiplicative update

```python
w = inference.fit_multiplicative(s[:l2],n,m)
```

```python
plt.scatter(w0,w)
plt.plot([-0.5,0.5],[-0.5,0.5],'r--')
mse = ((w0-w)**2).mean()
print('MSE by multiplicative update:',mse)
```
